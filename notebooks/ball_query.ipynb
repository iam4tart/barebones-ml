```python
import sys
sys.path.insert(0, "../")

import torch
import trimesh
import numpy as np
%matplotlib qt
import matplotlib.pyplot as plt

import barebones
from barebones import fps, ball_query

torch.manual_seed(42)
np.random.seed(42)
```

```python
K = 64
fps_result = fps.fps(pts, K)
fps_idx = fps_result[0]
fps_pts_tensor = fps_result[1]

print(f"fps centroids: {fps_pts_tensor.shape}")
```
